# K-Means Clustering Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv("clustering_dataset.csv")
X= np.matrix(df[['Distance_Feature','Speeding_Feature']])

kmeans = KMeans(n_clusters=2).fit(X)
kmeans.cluster_centers_
```

```python
kmeans.predict([[  181.24,   28.  ]])
```

```python
f1 = df['Distance_Feature'].values
f2 = df['Speeding_Feature'].values

plt.figure()
h1,=plt.plot(f1[kmeans.labels_==0],f2[kmeans.labels_==0],'go')
h2,=plt.plot(f1[kmeans.labels_==1],f2[kmeans.labels_==1],'bo')
plt.ylabel('Speeding Feature')
plt.xlabel('Distance Feature')
plt.legend([h1,h2],['Group 1','Group 2'], loc='upper left')

plt.plot(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],"ro")
plt.show()
```
